```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
sns.set_style("whitegrid")
```

```python
df = pd.read_csv("datas/market_fact.csv")
```

```python
df
```

```python
# Simple density plot
sns.distplot(df['Shipping_Cost'])
plt.show()
```

```python
sns.distplot(df['Shipping_Cost'][:200], rug=True)
plt.show()
```

```python
sns.distplot(df['Sales'], hist = False)
plt.show()
```

```python
plt.subplot(2, 2, 1)
plt.title('Sales')
sns.distplot(df['Sales'])

plt.subplot(2, 2, 2)
plt.title('Profit')
sns.distplot(df['Profit'])

plt.subplot(2, 2, 3)
sns.distplot(df['Order_Quantity'])

plt.subplot(2, 2, 4)
sns.distplot(df['Shipping_Cost'])
```

```python
# Boxplots

sns.boxplot(df['Order_Quantity'])
plt.title('Order_Quantity')
plt.show()
```

```python
sns.boxplot(y = df['Order_Quantity'])
plt.title('Order_Quantity')
plt.show()
```

```python
sns.jointplot('Sales', 'Profit', df)
plt.show()
```

```python
df = df[(df.Profit < 10000) & (df.Sales < 20000)]
sns.jointplot('Sales', 'Profit', df)
plt.show()
```
