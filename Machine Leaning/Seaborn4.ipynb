```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="white")
```

```python
market = pd.read_csv("datas/market_fact.csv")
orders = pd.read_csv("datas/orders_dimen.csv")
prod = pd.read_csv("datas/prod_dimen.csv")
shipping = pd.read_csv("datas/shipping_dimen.csv")
```

```python
sns.boxplot(y=market['Sales'])
plt.yscale('log')
plt.show()
```

```python
df = pd.merge(market, prod, how = 'inner', on = 'Prod_id')
df.head()
```

```python
sns.barplot(x = 'Product_Category', y = 'Sales', data = df)
plt.show()
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.barplot(x = 'Product_Category', y = 'Sales', data = df)
plt.title("Average_Sales")

plt.subplot(1, 2, 2)
sns.barplot(x = 'Product_Category', y = 'Sales', data = df, estimator = np.median)
plt.title("Median_Sales")

plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.barplot(x='Profit', y='Product_Sub_Category', data = df, estimator = np.median)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.countplot(y='Product_Sub_Category', data=df)
plt.show()
```
