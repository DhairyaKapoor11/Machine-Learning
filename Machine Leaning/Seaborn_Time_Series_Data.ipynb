```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="white")
```

```python
market = pd.read_csv("datas/market_fact.csv")
orders = pd.read_csv("datas/orders_dimen.csv")
prod = pd.read_csv("datas/prod_dimen.csv")
shipping = pd.read_csv("datas/shipping_dimen.csv")
```

```python
market.head()
```

```python
df = pd.merge(market, orders, how = 'inner', on = 'Ord_id')
df.head()
```

```python
df.info()
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df.info()
```

```python
time_df = df.groupby('Order_Date')['Sales'].sum()
print(time_df.head())
print(type(time_df))
```

```python
plt.figure(figsize=(16, 8))

sns.tsplot(data = time_df)
plt.show()
```

```python
df['month'] = df['Order_Date'].dt.month

df['year'] = df['Order_Date'].dt.year
```

```python
df.head()
```

```python
time_df = df.groupby(["year", "month"]).Sales.mean()
print(time_df.head())
```

```python
plt.figure(figsize=(8, 6))
sns.tsplot(time_df)
plt.xlabel("Time")
plt.ylabel("Sales")
plt.show()
```

```python
year_month = pd.pivot_table(df, values = 'Sales', index = 'year', columns = 'month')
print(year_month.head())
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(year_month, cmap="YlGnBu")
plt.show()
```
