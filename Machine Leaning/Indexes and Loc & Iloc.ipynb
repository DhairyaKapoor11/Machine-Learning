```python
import numpy as np
```

```python
temp = np.random.randint(low = 20, high = 100, size = [20,])
name = np.random.choice(['Dhairya', 'Meena', 'Riya', 'Mani'], 20)
```

```python
random = np.random.choice([1, 2, 3, 4], 20)
```

```python
x = list(zip(temp, name, random))
```

```python
import pandas as pd
df = pd.DataFrame(data = x, columns = ['temp', 'name', 'random'])
```
