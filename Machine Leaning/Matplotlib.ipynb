```python
# Change the colors & Line type

x = np.linspace(0, 5, 10)
y = np.linspace(3, 6, 10)

plt.plot(x, y, 'r-', x, y**2, 'b+', x, y**3, 'g^')
plt.show()
```

```python
x = np.linspace(1, 10, 100)
y = np.log(x)

plt.figure(1)

# First subplot
plt.subplot(121)
plt.title("y = log(x)")
plt.plot(x, y)

# Second Subplot
plt.subplot(122)
plt.title("y = log(x)**2")
plt.plot(x, y**2)
plt.show()
```

```python
x = np.linspace(1, 10, 100)

plt.figure(1)

plt.subplot(2, 2, 1)
plt.title("Linear")
plt.plot(x, x)

plt.subplot(2, 2, 2)
plt.title("Cubic")
plt.plot(x, x**3)

plt.figure(2)

plt.subplot(2, 2, 1)
plt.title("Log")
plt.plot(x, np.log(x))

plt.subplot(2, 2, 2)
plt.title("Exponential")
plt.plot(x, x**2)

plt.show()
```
